```python
# ACC102 Track2 GitHub Project: S&P 500 Stock Analysis
# Author: [Zixuan.Wang]
# Student ID: [2472236]

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# --------------------------
# 1. Data Generation (Simulated S&P 500)
# --------------------------
start_date = datetime(2023, 1, 1)
end_date = datetime(2026, 4, 10)
dates = pd.date_range(start=start_date, end=end_date, freq='B')

np.random.seed(42)
base_price = 3800
trend = np.linspace(0, 1500, len(dates))
volatility = np.random.normal(0, 30, len(dates))
prices = base_price + trend + volatility

data = pd.DataFrame({
    'Date': dates,
    'Open': prices + np.random.normal(0, 10, len(dates)),
    'High': prices + np.random.normal(10, 5, len(dates)),
    'Low': prices - np.random.normal(10, 5, len(dates)),
    'Close': prices,
    'Volume': np.random.randint(2000000, 5000000, len(dates))
})
data = data.set_index('Date')

# --------------------------
# 2. Data Cleaning
# --------------------------
data = data.dropna()

# --------------------------
# 3. Calculate Daily Return
# --------------------------
data['Daily_Return'] = data['Close'].pct_change()

# --------------------------
# 4. Descriptive Statistics
# --------------------------
print("===== Descriptive Statistics =====")
print(data[['Close', 'Daily_Return']].describe())

# --------------------------
# 5. Visualization
# --------------------------
plt.figure(figsize=(12, 5))
plt.plot(data['Close'], color='#1f77b4', label='S&P 500 Close Price')
plt.title('S&P 500 Daily Close Price (2023–2026, Simulated)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --------------------------
# 6. Key Insights
# --------------------------
max_price = data['Close'].max()
min_price = data['Close'].min()
avg_return = data['Daily_Return'].mean()

print("\n===== Key Insights =====")
print(f"Highest Closing Price: ${max_price:.2f}")
print(f"Lowest Closing Price: ${min_price:.2f}")
print(f"Average Daily Return: {avg_return:.4%}")
```
